#### Importing the libraries

```python
import pandas as pd
from googletrans import Translator
import re
import string
```

```python
# pip install googletrans
# pip install pandas
```

```python
translator = Translator()
```

1. The function below reads a JSON file as a dataframe.

2. Extracts requirement and salary field and removes empty field and punctuations.

3. Translates from Deutsch to English.

4. Then returns the cleaned fields.

```python
def data_cleaning_part_one(json_file):
    raw_df = pd.read_json(json_file)
    requirements = raw_df['requirement']
    salary_range = raw_df['salary']
    requirements = [req for req in requirements if req != [[]]]
    requirements = [req for reqs in requirements for req in reqs]
    requirements = [req for reqs in requirements for req in reqs]
    requirements = [translator.translate(text = req, dest = 'en').text for req in requirements]
    requirements = [req.translate(str.maketrans('', '', string.punctuation)).strip().replace('','') for req in requirements if req]
    salary_range.dropna(inplace=True)
    salary_range = [translator.translate(text=sal, dest="en").text.split() for sal in salary_range]
    salaries = [float(salary_range[i][5].split('€')[1].replace(',','')) for i in range(0, len(salary_range))]
    salaries = [float(salary_range[i][5].split('€')[1].replace(',','')) for i in range(0, len(salary_range))]
    salaries = [ sal*12 if len(str(sal)) == 4 else sal for sal in salaries]
    data = (salaries, requirements)
    return data
```

```python
result = data_cleaning_part_one('../Baden-Württemberg.json')
salaries = result[0]
requirements = result[1]
```

#### Converting to dataframe

```python
salary = pd.DataFrame(salaries, columns=['Salaries'])
salary
```

```python
requirements = pd.DataFrame(requirements, columns=['Requirements'])
requirements
```

```python
salary.to_csv("1.2 Baden-Württemberg_salaries.csv", index=False)
requirements.to_csv("1.1 Baden-Württemberg_requirements.csv", index=False)
```
